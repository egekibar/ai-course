```python
import spacy
from spacy import displacy
NER = spacy.load('en_core_web_sm')
```

```python
text='''
Zafer Acar loves NLP and artificial intelligence. Python is a great programming language.
The Turkish Space Agency is the national space agency of Turkey, headquartered in Ankara.
It operates under the Department of Space which is directly overseen
by the President of Turkey and administor of TUA is Serdar Hüseyin Yildirm.
'''
```

```python
for word in NER(text).ents:
    print(word.text, word.label_)
```

```python
spacy.explain('ORG')
```

```python
spacy.explain('PERSON')
```

```python
spacy.explain('GPE')
```

```python
displacy.render(NER(text))
```

```python
displacy.render(NER(text), style='ent', jupyter=True)
```
