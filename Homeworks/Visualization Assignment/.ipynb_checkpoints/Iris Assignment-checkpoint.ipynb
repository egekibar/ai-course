# "Human Learning" with iris data

Can you predict the species of an iris using petal and sepal measurements?

TASKS:

2. Gather some basic information about the data.
3. Use groupby, sorting, and/or plotting to look for differences between species.
4. Come up with a set of rules that could be used to predict species based upon measurements.

BONUS: Define a function that accepts a row of data and returns a predicted species.
Then, use that function to make predictions for all existing rows of data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Task 3

```python
# use groupby to look for differences between the species
mean_group = df.groupby('species').mean()
max_group = df.groupby('species').max()
min_group = df.groupby('species').min()

print('group by mean: ', mean_group)
print('group by max: ', max_group)
print('group by min: ', min_group)
```

```python
# use sorting to look for differences between the species
sorted_sepal_length = df.sort_values(by=['species', 'sepal_length'])
sorted_petal_length = df.sort_values(by=['species', 'petal_length'])

print("Sorted by Sepal Length:\n", sorted_sepal_length)
print("\nSorted by Petal Length:\n", sorted_petal_length)
```

```python
# use plotting to look for differences between the species
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df, palette='deep')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df, palette='deep')
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Sepal Length by Species')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_width', data=df)
plt.title('Sepal Width by Species')
plt.show()
```

```python
sns.pairplot(df, hue='species', palette='deep')
plt.show()
```

```python
# map species to a numeric value so that plots can be colored by category
species_to_numeric = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
df['species_num'] = df['species'].map(species_to_numeric)
plt.figure(figsize=(10, 6))
plt.scatter(df['sepal_length'], df['sepal_width'], c=df['species_num'], cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.colorbar(ticks=[0, 1, 2], format=plt.FuncFormatter(lambda val, loc: ['setosa', 'versicolor', 'virginica'][int(val)]))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['petal_length'], df['petal_width'], c=df['species_num'], cmap='viridis')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.title('Petal Length vs Petal Width')
plt.colorbar(ticks=[0, 1, 2], format=plt.FuncFormatter(lambda val, loc: ['setosa', 'versicolor', 'virginica'][int(val)]))
plt.show()
```

## Task 4

Setosa: Petal length < 2.0 <br/>
Versicolor: Petal length between 2.0 and 5.0, and petal width < 1.8 <br/>
Virginica: Petal length > 5.0 <br/>

## Bonus

```python
def predict_species(row):
    if row['petal_length'] <= 1.9:
        return 'setosa'
    elif row['petal_length'] >= 5 or row['petal_width'] >= 1.7:
        return 'virginica'
        
    elif row['petal_length'] < 5 or row['petal_width'] < 1.7:
        return 'versicolor'
    
        


df['predicted_species'] = df.apply(predict_species, axis=1)

df.sample(20)
```

```python
setosa = df[df.species == 'setosa']
versicolor = df[df.species == 'versicolor']
virginica = df[df.species == 'virginica']


versicolor_sortby_petal_width = versicolor.sort_values(by='petal_width', ascending=False)

versicolor_sortby_petal_lenght = versicolor.sort_values(by='petal_length', ascending=False)
 
missing_versicolor = len(versicolor[versicolor['species'] != versicolor['predicted_species']])

accuracy_setosa = (setosa['species'] == setosa['predicted_species']).mean()
accuracy_versicolor = (versicolor['species'] == versicolor['predicted_species']).mean()
accuracy_virginica = (virginica['species'] == virginica['predicted_species']).mean()

accuracy = (df['species'] == df['predicted_species']).mean()


print(f"General Prediction Accuracy: {accuracy * 100:.2f}%")

print(f"setosa Prediction Accuracy: {accuracy_setosa * 100:.2f}%")

print(f"versicolor Prediction Accuracy: {accuracy_versicolor * 100:.2f}%")

print(f"virginica Prediction Accuracy: {accuracy_virginica * 100:.2f}%")

missing_versicolor
```
