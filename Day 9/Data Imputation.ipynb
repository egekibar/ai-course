Boş verileri neyle doldurabiliriz? (Rakamsa)

- **Linear Regression**
- Medium ( Ortalama )
- Median ( Orta değer )
- Mod ( En çok tekrar eden )
- KNNImputer ( En yakın komşuyla )

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
imp = IterativeImputer()
imputed_data = imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data)
```

```python
from sklearn.impute import KNNImputer
knn = KNNImputer()
```

```python
imputed_data2 = knn.fit_transform(df)
```

```python
pd.DataFrame(imputed_data2, columns = df.columns)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp = SimpleImputer(strategy='mean') # median, mode
```

```python
imputed_data3 = imp.fit_transform(df)
```

```python
pd.DataFrame(imputed_data3, columns = df.columns)
```
