# Scaling - Normalization - PCA

```python
import pandas as pd
```

```python
df = pd.read_csv('pima-indians-diabetes.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import normalize, scale # z score
```

```python
x = df.drop('Outcome', axis=1)
```

```python
x.head()
```

```python
yenix = scale(x)
```

```python
pd.DataFrame(yenix)
```

```python
normalizedata = normalize(x)
```

```python
pd.DataFrame(normalizedata)
```

### Boyut Küçültme - PCA ( Priciple Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
```

```python
x.head()
```

```python
data2 = pca.fit_transform(x)
```

```python
data2
```

```python
pd.DataFrame(data2[:, :2], columns=['PCA1', 'PCA2'])
```

```python
sum(pca.explained_variance_ratio_)
```
